```python
ndf.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X=ndf.drop(columns=['diagnosis'])
y=ndf['diagnosis']

X_train , X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train.head()
```

```python
y_train = pd.Series(y_train)
```

```python
X_train.info()
```

```python
num_vars = X_train.select_dtypes(include='number').columns
```

```python
stdsc = StandardScaler()

X_train[num_vars] = stdsc.fit_transform(X_train[num_vars])
X_test[num_vars] = stdsc.transform(X_test[num_vars])
```

```python
X_train.head()
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import KFold, cross_val_score, cross_val_predict
```

```python
kf = KFold(shuffle=True, random_state=42)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
def k_cross_val(model , X_train, y_train , cv):
    
    scores = cross_val_score(estimator = model, X=X_train, y=y_train, cv=cv,scoring='accuracy', n_jobs=-1)

    print(f'Accuracy Scores for {model} are : {scores}\n')
    print(f'Mean Accuracy score : {scores.mean()}\n')


    y_pred = cross_val_predict(estimator = model, X=X_train, y=y_train, cv=cv, n_jobs=-1)

    print(f'Accuracy of Predictions : {accuracy_score(y_train, y_pred)}\n')
    print(f'Classification Report of Predictions : {classification_report(y_train, y_pred)}\n')
    print(f'Confusion Matrix of Predictions : \n{confusion_matrix(y_train, y_pred)}')
```

```python
k_cross_val(LogisticRegression(), X_train, y_train, cv=kf)
```

```python
k_cross_val(SVC(), X_train, y_train, cv=kf)
```

```python
k_cross_val(RandomForestClassifier(), X_train, y_train, cv=kf)
```

## We got the least amount of error (typ1,typ2) in RandomForestClassifier
So we go ahead with RandomForest

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_param = {
    'max_depth' : [None, 5, 8, 10, 15, 20],
    'n_estimators' : [50,100,200],
    'criterion' : ['gini','entropy'],
    'max_features' : ['sqrt','log2'],
} 

```

```python
model = RandomForestClassifier(n_jobs=-1, random_state=42)
```

```python
gc = GridSearchCV(estimator=model, param_grid=grid_param, scoring = 'accuracy' , cv=kf )
```

```python
gc.fit(X_train, y_train)
```

```python
gc.best_estimator_
```

```python
gc.best_params_
```

```python
model = gc.best_estimator_
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
acc = accuracy_score(y_test,y_pred)
rep = classification_report(y_test,y_pred)
cm= confusion_matrix(y_test,y_pred)

print(f'Final Accuracy Score : {acc}')
print(f'Final Report :\n{rep}\n')
print(f'Final Confusion Matrix :\n{cm}')

sns.heatmap(data=cm, annot=True)
```
