```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Categorical Data Plots

These are for qualitative data

1) boxplot
2) violinplot
3) stripplot
4) swarmplot
5) barplot
6) countplot

```python
df = sns.load_dataset('tips')
df.head()
```

## Countplot 

Shows frequency of categories

No numeric variable required

Class distribution

Target imbalance checking

```python
sns.countplot(df['sex'])
```

```python
# using other col data for comparing using hue

sns.countplot(x = df['sex'], hue = df['smoker'])
```

## Bar plot

Displays aggregated statistics (mean by default)

Includes confidence intervals

x - qualitative data (Continous)

y - quantitative data (Categorical)

```python
sns.barplot(x = df['sex'],y = df['total_bill'])

# on y-axis . it shows mean of numeric data to compare not actuall values
```

```python
# to see total actuall values in Y, we use .sum as estimator

sns.barplot(x = df['sex'],y = df['total_bill'] , estimator = np.sum) # estimator is how Y is scaled in plot,default is mean
```

## Box plot

Uses 5 number summary

Summarizes a numeric variable grouped by categories

Shows:

Median (Q2)

Quartiles (Q1, Q3)

Interquartile Range (IQR)

Outliers

How does a numeric feature vary across categories?

Are there outliers in any category?

```python
sns.boxplot(x='tip', y='day', data=df)
```

```python
sns.boxplot(x='tip', hue='sex', data=df,palette = 'rainbow') # hue is for categorical data..can still work with y, but will be deprecated when palette is assigned
```

boxplots and violinplots are used to shown the distribution of categorical data. A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be "outliers" using a method that is a function of the inter-quartile range.

## Violin Plot

similar to box plot

Combines boxplot + KDE

```python
sns.violinplot(x='day',y = 'total_bill',data=df)
```

```python
sns.violinplot(x='tip',hue = 'sex',data=df,palette='rainbow')
```

## Stripplot

A stripplot draws a scatter plot where one variable is categorical and the other is numeric.

Displays all individual observations

Points are placed along the categorical axis

The stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

```python
sns.stripplot(x='day',y='total_bill',data=df)
```

```python
sns.stripplot(x='day',y='total_bill',data=df,palette='rainbow')
```

## Swarmplot

A swarmplot is similar to a stripplot, but points are adjusted to avoid overlap.

Displays all individual observations

Automatically arranges points along the categorical axis

Ensures no overlapping points

The swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they dont overlap. This gives a better representation of the distribution of values, although it does not scale well to large numbers of observations (both in terms of the ability to show all the points and in terms of the computation needed to arrange them).

```python
sns.swarmplot(x='day',y='total_bill',data=df)
```

```python
sns.violinplot(x='total_bill',y='day',data=df) # shows data distribution
sns.swarmplot(x='total_bill',y='day',data=df)  # shows individual data points
```
