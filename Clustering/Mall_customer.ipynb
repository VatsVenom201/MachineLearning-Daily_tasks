```python
X = df.drop(columns=['Preferred Category','Loyalty Years'],axis =1 )
y = df['Preferred Category']

X.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# to get value of K we compare using a loop

for k in range (2,11):
    km = KMeans(n_clusters=k,random_state=42)
    km.fit(X)
    labels = km.labels_
    sc = silhouette_score(X,labels)
    print(f'Value of K:{k}')
    print(f'Silhouette_score:{sc}')
```

### Through above we get K=4 is the best value

```python
km = KMeans(n_clusters=4,random_state=42)

km.fit(X)
```

```python
labels = km.labels_
```

## Evaluation of Clustering

2 types of eval: External & Internal

Internal : does need true labels (true labels VS predicted)

External : doesn't need true labels

#### Internal

Silhouette Score - Measures how similar a point is to its own cluster vs other clusters (range -1 to 1)(1 is best).

#### External

Homogeneity Score - Checks if each cluster contains only members of a single class.

Completeness Score - Checks if all members of a class are assigned to the same cluster.

V-Measure - Harmonic mean of homogeneity and completeness.

Adjusted Mutual Information (AMI) - Measures shared information between predicted clusters and true labels.

Adjusted Rand Index (ARI) - Measures similarity between predicted clusters and true labels, adjusted for chance.



```python
from sklearn.metrics import homogeneity_score, completeness_score, v_measure_score, adjusted_rand_score, adjusted_mutual_info_score
```

```python
# EXTERNAL
homog = homogeneity_score(y,labels)
comp = completeness_score(y,labels)
vm = v_measure_score(y,labels)
adjr = adjusted_rand_score(y,labels)
adjm = adjusted_mutual_info_score(y,labels)

# INTERNAL
sil= silhouette_score(X,labels)

print(f'homogeneity_score:{homog}')
print(f'completeness_score:{comp}')
print(f'v_measure_score:{vm}')
print(f'adjusted_rand_score:{adjr}')
print(f'adjusted_mutual_info_score:{adjm}')
print(f'Silhouette Score:{sil}')
```
